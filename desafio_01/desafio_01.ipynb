```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL da página inicial do JC UOL:
url = "https://jc.uol.com.br/"
response = requests.get(url)

# Criar o objeto BeautifulSoup para fazer o parsing do HTML
soup = BeautifulSoup(response.text, "html.parser")

# Encontrar todas as seções de manchete na página
hard_news_section = soup.find("section", id="hardnews1")
jc_360_section = soup.find("section", id="section-jc360")
videos_section = soup.find("section", id="section-videos")
pe_section = soup.find("section", id="section-pernambuco")

# Função para modularizar a extração de títulos e links das seções
def extract_titles_and_links(section):
  news_list = []
  anchors_tags = section.select("a:has(div.content-text)")
  for anchor_tag in anchors_tags:
      headline_tag = anchor_tag.find(["h2", "h4"])
      headline_title = headline_tag.text.strip() if headline_tag else None

      news_link = anchor_tag.get("href")

      news_list.append({
          "titulo": headline_title,
          "link": news_link
      })
  return news_list

# Extrair listas de cada seção
hard_news_list = extract_titles_and_links(hard_news_section)
jc_360_list = extract_titles_and_links(jc_360_section)
videos_list = extract_titles_and_links(videos_section)
pe_list = extract_titles_and_links(pe_section)

# Merge de todas as listas
all_news = hard_news_list + jc_360_list + videos_list + pe_list

# Imprime no console a lista de títulos de manchetes
for i, headline in enumerate(all_news, start=1):
    print(f'Notícia {i}:')
    print(f'- Título: {headline["titulo"]}')
    print(f'- Link: {headline["link"]}\n')

# Criar um DataFrame com os dados e salvar em um arquivo CSV
df_headlines = pd.DataFrame(all_news)
print("\n\nDataFrame:")
print(df_headlines)
df_headlines.to_csv("lista_de_noticias.csv", index=False)

print("\nDetalhes das notícias salvos em lista_de_noticias.csv")
```
